# ParagonCorp Products Sales Quantities Forecasting

# Import Library

```python
# Import Library
import pickle
import pandas as pd

# Visualization Library
import matplotlib.pyplot as plt
import seaborn as sns
```

# Model Loading

```python
# Model Loading 
with open('model_sarimax.pkl', 'rb') as file_1:
  model_sarimax = pickle.load(file_1)
```

# Forecasting

```python
# Forecasting
model_sarimax.forecast(18).tail(5)
```

```python
# Forecast Visualization
plt.figure(figsize=(20,10))
result = pd.DataFrame(model_sarimax.forecast(18).tail(5))
sns.lineplot(x=result.index, y=result.predicted_mean, data=result)
plt.title('Prediction for the next 5 weeks', fontsize=20)
plt.xlabel('Date', fontsize=16)
plt.xticks(fontsize=8)
plt.ylabel('Quantities Difference', fontsize=16)
plt.show()
```
